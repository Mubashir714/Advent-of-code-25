```python
def simulate_beams(grid):
    rows, cols = len(grid), len(grid[0])
    
    # Find starting position S
    start_y, start_x = next((i, j) for i in range(rows) for j in range(cols) if grid[i][j] == 'S')
    
    # Directions: only downward propagation
    directions = [(1, 0)]  # down
    
    splits = 0
    visited = set()
    
    # BFS queue: (y, x)
    from collections import deque
    queue = deque([(start_y, start_x)])
    visited.add((start_y, start_x))
    
    while queue:
        y, x = queue.popleft()
        
        # Move down one step
        ny, nx = y + 1, x
        
        # Out of bounds = exit (no split)
        if ny >= rows:
            continue
            
        cell = grid[ny][nx]
        
        if cell == '^':
            # SPLIT! Count +1, emit left and right beams
            splits += 1
            
            # Left beam
            if nx - 1 >= 0 and (ny, nx - 1) not in visited:
                queue.append((ny, nx - 1))
                visited.add((ny, nx - 1))
            
            # Right beam  
            if nx + 1 < cols and (ny, nx + 1) not in visited:
                queue.append((ny, nx + 1))
                visited.add((ny, nx + 1))
        elif cell == '.':
            # Continue straight down
            if (ny, nx) not in visited:
                queue.append((ny, nx))
                visited.add((ny, nx))
    
    return splits

def main():
    # Parse your exact input grid
    grid = []
    with open("input.txt", "r") as f:
        for line in f:
            grid.append(list(line.rstrip()))
    
    result = simulate_beams(grid)
    print(f"Part 1: {result}")
    
    with open("output.txt", "w") as f:
        f.write(str(result) + "\n")
    
    print("Result written to output.txt")

if __name__ == "__main__":
    main()
```
