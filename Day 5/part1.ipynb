```python
with open('input.txt', 'r') as f:
    lines = [line.strip() for line in f.readlines() if line.strip()]

ranges = []
available = []

# Parse ranges (first section until blank line implied by format)
i = 0
while i < len(lines) and '-' in lines[i]:
    start, end = map(int, lines[i].split('-'))
    ranges.append((start, end))
    i += 1

# Parse available IDs (remaining lines)
while i < len(lines):
    available.append(int(lines[i]))
    i += 1

# Count fresh ingredients optimally
fresh_count = 0
for id_val in available:
    if any(start <= id_val <= end for start, end in ranges):
        fresh_count += 1

with open('output.txt', 'w') as f:
    f.write(str(fresh_count))
```
