```python
def merge_ranges(ranges):
    if not ranges:
        return []
    ranges.sort()
    merged = [list(ranges[0])]
    for current in ranges[1:]:
        if merged[-1][1] >= current[0] - 1:
            merged[-1][1] = max(merged[-1][1], current[1])
        else:
            merged.append(list(current))
    return merged

with open('input.txt', 'r') as f:
    lines = [line.strip() for line in f.readlines() if line.strip() and '-' in line]

ranges = [tuple(map(int, line.split('-'))) for line in lines]

merged_ranges = merge_ranges(ranges)
total_fresh = sum(end - start + 1 for start, end in merged_ranges)

with open('output.txt', 'w') as f:
    f.write(str(total_fresh))
```
