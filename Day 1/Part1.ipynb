```python
# Advent of Code 2025 - Day 1 Part 1
# Reads rotations from input.txt
# Writes the number of times the dial lands on 0 into output.txt

def solve():
    current = 50   # starting position
    count_zero = 0

    with open("input.txt", "r") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            direction = line[0]
            value = int(line[1:]) % 100  # reduce large steps modulo 100

            if direction == 'R':
                current = (current + value) % 100
            else:  # direction == 'L'
                current = (current - value) % 100

            if current == 0:
                count_zero += 1

    with open("output.txt", "w") as f:
        f.write(str(count_zero))


if __name__ == "__main__":
    solve()
```
