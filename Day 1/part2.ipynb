```python
def main():
    position = 50  # starting point
    zero_count = 0

    with open("input.txt", "r") as file:
        for line in file:
            line = line.strip()
            if not line:
                continue
            direction = line[0]
            distance = int(line[1:])

            if direction == 'L':
                # Left rotation: count every step that hits 0
                for _ in range(distance):
                    position = (position - 1) % 100
                    if position == 0:
                        zero_count += 1
            elif direction == 'R':
                # Right rotation: count every step that hits 0
                for _ in range(distance):
                    position = (position + 1) % 100
                    if position == 0:
                        zero_count += 1

    with open("output.txt", "w") as outfile:
        outfile.write(str(zero_count) + "\n")


if __name__ == "__main__":
    main()
```
