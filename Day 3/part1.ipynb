```python
def max_joltage(bank):
    """Find max 2-digit joltage from any two positions in bank"""
    max_val = 0
    n = len(bank)
    
    for i in range(n):
        for j in range(i + 1, n):
            joltage = int(bank[i] + bank[j])
            if joltage > max_val:
                max_val = joltage
    return max_val

def main():
    total_joltage = 0
    
    with open("input.txt", "r") as file:
        for line in file:
            bank = line.strip()
            if bank:  # Skip empty lines
                total_joltage += max_joltage(bank)
    
    with open("output.txt", "w") as outfile:
        outfile.write(str(total_joltage) + "\n")

if __name__ == "__main__":
    main()
```
