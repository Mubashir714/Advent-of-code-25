```python
def largest_12_digit_number(s):
    """Remove (len(s)-12) digits to get largest 12-digit number preserving order"""
    if len(s) < 12:
        return 0
    
    stack = []
    digits_to_remove = len(s) - 12
    
    for digit in s:
        # While we can remove smaller digits to make room for this larger one
        while stack and digits_to_remove > 0 and stack[-1] < digit:
            stack.pop()
            digits_to_remove -= 1
        stack.append(digit)
    
    # Remove remaining digits from end if needed
    while digits_to_remove > 0:
        stack.pop()
        digits_to_remove -= 1
    
    # Convert to number (remove leading zeros if any, but shouldn't happen)
    result = ''.join(stack)
    return int(result)

def main():
    total_joltage = 0
    
    with open("input.txt", "r") as f:
        for line in f:
            bank = line.strip()
            if bank:
                total_joltage += largest_12_digit_number(bank)
    
    print(total_joltage)
    with open("output.txt", "w") as f:
        f.write(str(total_joltage) + "\n")

if __name__ == "__main__":
    main()
```
