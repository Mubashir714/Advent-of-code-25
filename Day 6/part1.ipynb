```python
def solve():
    # Read all lines from input.txt
    with open("input.txt", "r") as f:
        lines = [line.rstrip("\n") for line in f.readlines()]

    # Remove trailing empty lines (just in case)
    while lines and not lines[-1].strip():
        lines.pop()

    # Last line is the operator row
    ops_line = lines[-1]
    number_lines = lines[:-1]

    width = len(ops_line)

    # Mark columns that are fully empty (spaces in all number rows)
    is_space = [
        all((row[i] == " " if i < len(row) else True) for row in number_lines)
        for i in range(width)
    ]

    # Group columns into problems separated by space columns
    problem_columns = []
    current = []
    for i, space in enumerate(is_space):
        if space:
            if current:
                problem_columns.append(current)
                current = []
        else:
            current.append(i)

    if current:
        problem_columns.append(current)

    grand_total = 0

    # Process each problem block
    for cols in problem_columns:
        numbers = []

        # Extract numbers from vertical slices
        for row in number_lines:
            chunk = "".join(row[i] if i < len(row) else " " for i in cols).strip()
            if chunk:
                numbers.append(int(chunk))

        # Determine operator
        op = ops_line[cols[0]]

        # Compute block value
        if op == '+':
            val = sum(numbers)
        else:
            val = 1
            for n in numbers:
                val *= n

        grand_total += val

    # Write answer to output.txt
    with open("output.txt", "w") as f:
        f.write(str(grand_total))


if __name__ == "__main__":
    solve()
```
